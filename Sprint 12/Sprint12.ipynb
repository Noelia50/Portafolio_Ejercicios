2. Mostra per pantalla:

- La quantitat total de cada recurs.
- El codi d'estat de cada petició.

```python
elementos = ["posts", "comments", "users"]

for elem in elementos:
    x = requests.get(f"https://jsonplaceholder.typicode.com/{elem}")
    if x.status_code == 200:
        print(f"Quantitat de {elem}: {len(x.json())}")
        print(f"Codi d'estat de {elem}: {x.status_code}\n")
```

4. Fes una petició POST per crear una nova publicació fictícia. Inclou el títol, el cos del missatge i un userId.

- Mostra la resposta JSON.
- Mostra el codi d'estat.

```python
url = "https://jsonplaceholder.typicode.com/posts"

data = {
    "title": "Este es mi post",
    "body": "Este es el contenido del post",
    "userId": 1
}

response = requests.post(url, json=data)

print(response.json()) 
print(response.status_code)
```

Nota personal:
- POST → NO envías id
- PUT/PATCH → SÍ usas id (en la URL)


POST   /posts        → crear (sin id)

PATCH  /posts/99     → modificar id=99

DELETE /posts/99     → eliminar id=99

5. Fes una petició PATCH per modificar parcialment una publicació existent.

- Mostra la resposta JSON.
- Mostra el codi d'estat.

```python
url = "https://jsonplaceholder.typicode.com/posts/1"

data = {
    "title": "Título actualizado"
}

response = requests.patch(url, json=data)

print(response.json())
print(response.status_code)
```

6. Fes una petició DELETE sobre una publicació.

- Mostra la resposta JSON.
- Mostra el codi d'estat.

```python
url = "https://jsonplaceholder.typicode.com/posts/1"

response = requests.delete(url)

print(response.json())
print(response.status_code)
```

La funció ha de retornar un diccionari amb les següents dades relacionades amb el continent indicat:

Nombre de regions diferents que hi ha dins del continent.
Nombre de països dins del continent.
Superfície màxima (SurfaceArea) registrada al continent.

Població mitjana (Population) dels països del continent.
Mitjana de l’esperança de vida (LifeExpectancy) dels països del continent.

Finalment, carrega el fitxer adjunt amb la següent instrucció:

data = pd.read_csv("arxiu csv", sep=";", encoding="latin-1")

Para calcular el IMC es necesario dividir el peso en kg de un paciente entre el cuadrado de su altura en metros

Bajo peso: menos de 18,5.
Peso normal: 18,5-24,9.
Sobrepeso: 25-29,9.

### **2. Convertidor de temperatures**

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

```python
def convertir_temperatura(valor, t_origen):

    conversiones = {
        "C": { #diccionario de funciones anonimas y compactas
            "F": lambda x: (x * 1.8) + 32, #t_destino: funcion y funcion: valor, siendo valor el parametro de la funcion anonima y se define al llamr a la funcion convertir_temperatura
            "K": lambda x: x + 273.15
        },
        "F": {
            "C": lambda x: (x - 32) / 1.8,
            "K": lambda x: (x - 32) / 1.8 + 273.15
        },
        "K": {
            "C": lambda x: x - 273.15,
            "F": lambda x: (x - 273.15) * 1.8 + 32
        }
    }

    resultados = {}

    if t_origen not in conversiones:
        raise ValueError("Unidad no válida. Usa C, F o K.")

    for t_destino, funcion in conversiones[t_origen].items():
        resultados[t_destino] = round(funcion(valor), 2)

    return resultados

#validamos formato valor de temp y unidad temperatura de origen
while True: #lo primero que se ejecutara es el bucle
    try:
        temp = float(input("Introduce la temperatura: ")) #valor
        unidad = input("Unidad (C, F, K): ").upper() #origen

        resultado = convertir_temperatura(temp, unidad)
        break #si pasa la validación, guarda la info en resultado y rompe el bucle

    except ValueError as e:
        print(f"Error: {e}") #si no pasa la validación devolvera el error hasta que este bien


print(f"Para la temperatura {temp} {unidad}, sus conversiones son:", resultado)
```

### **3. Comptador de paraules d'un text.**

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
def cuenta_palabras(texto):
    recuento = {}
    texto = texto.lower()
    palabras = texto.split()
    signos = '.,;:!?()[]{}"\''

    #contadores
    palabras_acumuladas = 0
    letras_acumuladas = 0

    for palabra in palabras:
        palabra_correcta = palabra.strip(signos)
        if palabra_correcta: #evitamos palabras huecas
            palabras_acumuladas += 1 
            letras_acumuladas += len(palabra_correcta)
            if palabra_correcta in recuento:
                recuento[palabra_correcta] += 1
            else:
                recuento[palabra_correcta] = 1
    
    media_palabras = round((letras_acumuladas/palabras_acumuladas), 2) if palabras_acumuladas > 0 else 0 #operador ternario: versión reducida del if/else
    return recuento, media_palabras #esto devuelve un tupla como resultado[0] y resultado[1]

palabras_contadas = cuenta_palabras(input('Escriba una frase a continuación: '))
recuento, media_palabras = palabras_contadas
print(f'Recuento de las palabras: {recuento}')
print(f'Longitud media de las las palabras: {media_palabras}')
```

### **4. Diccionari invers (amb possibilitat de duplicats)**

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

Fixa’t en el següents exemples per veure l’output esperat.

```python
def diccionario_invertido(diccionario):
    invertido = {}
    hay_duplicados = False  # bandera para avisar

    for clave, valor in diccionario.items(): #importante indicar .items() sino no podemos iterar en pares sino en claves y nos dara error.
        if valor in invertido:
            hay_duplicados = True  # check: hay duplicado
            # El str se convierte a lista si aún no lo es, para poder hacer append
            if isinstance(invertido[valor], list) is False:
                invertido[valor] = [invertido[valor]] # [] 
            invertido[valor].append(clave)
        else:
            invertido[valor] = clave # no hay duplicados, solo se añade a la inversa

    if hay_duplicados:
        print("Existen valores duplicados en el diccionario invertido.")
        for valor, claves in invertido.items():
            if isinstance(claves, list):
                print(f"El valor '{valor}' está duplicado en las claves: {claves}")

    return invertido

resultado = diccionario_invertido({'x': 'apple', 'y': 'banana', 'z': 'banana' }) #diccionario de ejemplo
print(resultado)
```

### **2. Conversió de tipus de dades.**

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

```python
def conversion(lista): #estoy "aplanando y convirtiendo con una función"
    flotantes = []
    otros = []
    
    pendientes = list(lista) #copia a modificar, buenas practicas

    while pendientes:
        elemento = pendientes.pop(0) #sacamos elementos y usamos uno a uno. pop() elimina y devuelve el ultimo elemento

        if isinstance(elemento, (list, tuple)):
            pendientes.extend(elemento)  #a diferencia de append, extend desglosa el iterable y lo añade uno a uno
        else:
            try:
                numero = float(elemento)
                flotantes.append(numero)
            except (ValueError, TypeError):
                otros.append(elemento)

    return flotantes, otros

conversion([ '1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1,1.4, 'not-a-number'), [1,2,'3', '3.4']])
```

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

- longitud (int): Longitud de la contrasenya
- majuscules (bool = True): Si hi ha d’aparèixer majúscules
- minuscules (bool = True): Si hi ha d’aparèixer minúscules
- numeros (bool = True): Si hi ha d’aparèixer números
- signes (bool = False): Si hi ha d’aparèixer caràcters especials (,-$? o similars)

Així doncs, si executem la funció de la següent manera:

Hauríem d’obtenir un output (que hem de poder guardar) de l’estil:

9Er,5Vn8P$

Assegura’t que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleatòries.

(EXTRA) Explora com podríem fer que la funció copiés la contrasenya automàticament al porta papers de l’ordinador (com si l’haguéssim seleccionat i fet ctrl+copy).

```python
import random
```

```python
import pyperclip #para ctrl+copy
```

```python
def crear_contraseña(long:int, may:bool=True, min:bool=True, num:bool=True, esp:bool=False) -> str:
    minusculas="abcdefghijklmnñopqrstuvwxyz"
    mayusculas=minusculas.upper()
    numeros="0123456789"
    especiales="@()[]{}*;.,¿?¡!€$&%#=+-<>"
    base=""
    if may:
        base+=mayusculas
    if min:
        base+=minusculas
    if num:
        base+=numeros
    if esp:
        base+=especiales
    

    muestra=random.sample(base,long) #nos devuelve una lista de elementos aleatoria
    contraseña="".join(muestra) #string vacio+join(objeto iterable), nos devuelve la muestra concatenada
    pyperclip.copy(contraseña)
    return contraseña

clave = crear_contraseña(10, True, True, True, True)
print(f'Clave: {clave} copiada en el portapapeles')
```

### **2. Processament de dades simple**

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Té un històric de partits de futbol català en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera automàtica, per extreure’n els resultats que necessita.

Utilitza el fitxer "historic_partits.txt"

Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l’equip més golejador.
El nom de l’equip més golejat
La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)
Com a referència, els equips amb més punts serien els següents:

```python
import pandas as pd
```

```python
partidos = []
with open("/Users/noeliasolano/python/txts/historic partits.txt", "r", encoding="utf-8") as archivo:
    for linea in archivo:
        linea = linea.strip()
        if not linea:
            continue
        equipo1, resultado, equipo2 = linea.split("\t")
        goles1, goles2 = map(int, resultado.split("-")) #map() aplica una función a cada elemento de un iterable --> int
        partidos.append((equipo1, goles1, goles2, equipo2))

equipos = set() #como una lista pero no permite elementos repetidos
for elemento in partidos:
    equipos.update([elemento[0], elemento[3]]) #añade nombres de equipos de estos partidos al set a no ser que ya estuviesen

df = pd.DataFrame({
    "Equipo": list(equipos),
    "Partidos": 0,
    "Goles a Favor": 0,
    "Goles en Contra": 0,
    "PTS": 0
})
df.set_index("Equipo", inplace=True) #la columna indice sera la del equipo

for equipo1, goles1, goles2, equipo2 in partidos:
    #Partidos
    df.loc[equipo1, "Partidos"] += 1
    df.loc[equipo2, "Partidos"] += 1
    #Goles
    df.loc[equipo1, "Goles a Favor"] += goles1
    df.loc[equipo1, "Goles en Contra"] += goles2
    df.loc[equipo2, "Goles a Favor"] += goles2
    df.loc[equipo2, "Goles en Contra"] += goles1
    #Puntos
    if goles1 > goles2:
        df.loc[equipo1, "PTS"] += 3
    elif goles2 > goles1:
        df.loc[equipo2, "PTS"] += 3
    else:
        df.loc[equipo1, "PTS"] += 1
        df.loc[equipo2, "PTS"] += 1

# Diferencia de goles
df["Diferencia"] = df["Goles a Favor"] - df["Goles en Contra"]

# Orden por puntos, diferencia de goles, por Goles a Favor
df = df.sort_values(by=["PTS", "Diferencia", "Goles a Favor"], ascending=False)

#Tabla
print("=== Classificación Global ===")
print(df)

#Equipo mas goleador y mas goleado
max_gols = df["Goles a Favor"].max()
mas_goleador = df[df["Goles a Favor"] == max_gols].index.tolist()

max_gols_rebuts = df["Goles en Contra"].max()
mas_goleado = df[df["Goles en Contra"] == max_gols_rebuts].index.tolist()

print("\nEquipo mas goleador:", mas_goleador)
print("Equipo mas goleado:", mas_goleado)
```

Nuestra empresa se dedica a realizar análisis de datos geográficos y sociales, y un cliente nos ha solicitado una herramienta para identificar datos básicos de cada continente de manera automática.

1- Por ello, es necesario crear una función que reciba dos parámetros:

- Un DataFrame con información de países (como el del fichero Country.xlsx).
- El nombre de un continente (como cadena de texto).

2- La función debe devolver un diccionario con los siguientes datos relacionados con el continente indicado:

- Número de regiones diferentes que hay dentro del continente.
- Número de países dentro del continente.
- Superficie máxima (SurfaceArea) registrada en el continente.
- Superficie mínima registrada.
- Población media (Population) de los países del continente.
- Promedio de la esperanza de vida (LifeExpectancy) de los países del continente.

3- Finalmente, carga el fichero adjunto con la siguiente instrucción:

datos = pd.read_csv("arxiu csv", sep=";", encoding="latin-1")


Ejecuta la función para varios continentes y muestra los resultados obtenidos.

```python
# activando las bibliotecas
import pandas as pd
```

```python
# Cargando el dataframe
df_country = pd.read_csv("Country(in).csv", sep=";", 
                         encoding="latin-1")

df_country
```

```python
#Creando la función

def resumen_continente(df, continente):
    df_filtrado = df[df['Continent'] == continente] #filtrando por "continente"
    resumen = {
        'Regiones distintas': df_filtrado['Region'].nunique(),
        'Número de países': df_filtrado['Name'].count(),
        'Superficie máxima': df_filtrado['SurfaceArea'].max(),
        'Superficie mínima': df_filtrado['SurfaceArea'].min(),
        'Población media': df_filtrado['Population'].mean(),
        'Esperanza de vida promedio': df_filtrado['LifeExpectancy'].mean()
    }
    return resumen
```

```python
# Evaluando que hay en la columna "Continent"
df_country['Continent'].unique()
```

```python
# Usando la función
resumen_continente(df_country, "Africa")
```

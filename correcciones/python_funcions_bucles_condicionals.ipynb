1) Crea una funció que, donada una frase o una paraula, retorni el nombre de vocals. Un cop definida, sol·licita una paraula o frase a l’usuari i imprimeix el nombre de vocals.

```python
# Definim una funció que compta el nombre de vocals dins d’un text
def comptar_vocals(text):
    # Creem una cadena amb totes les vocals (minúscules i majúscules, incloent accents i dièresi)
    vocals = "aeiouàèéíïòóúüAEIOUÀÈÉÍÏÒÓÚÜ"

    # Inicialitzem un comptador a 0
    comptador = 0

    # Recorrem cada lletra del text
    for lletra in text:
        # Si la lletra és una vocal, incrementem el comptador en 1
        if lletra in vocals:
            comptador += 1

    # Retornem el nombre total de vocals trobades
    return comptador

# Demanem a l’usuari que introdueixi una paraula o frase
entrada = input("Introdueix una paraula o una frase: ")

# Cridem la funció i guardem el resultat a la variable 'nombre_vocals'
nombre_vocals = comptar_vocals(entrada)

# Imprimim el resultat per pantalla
print(f"El nombre de vocals és: {nombre_vocals}")
```

2. Defineix una funció que, donat un número, imprimeixi la seva taula de multiplicar (Ex: 1 x 1 = 1, 1 x 2 = 2, 1 x 3 = 3, ...). 

```python
# Definim una funció que mostra la taula de multiplicar d’un número
def taula_multiplicar(numero):
    # Recorrem els números de l’1 al 10 (amb range(1, 11))
    for i in range(1, 11):
        # Calculem el resultat de multiplicar el número per l’índex actual
        resultat = numero * i
        # Imprimim el resultat en format "número x i = resultat"
        print(f"{numero} x {i} = {resultat}")

# Demanem a l’usuari que introdueixi un número (convertit a enter amb int)
entrada = int(input("Introdueix un número per veure la seva taula de multiplicar: "))

# Cridem la funció passant-li el número introduït per l’usuari
taula_multiplicar(entrada)
```

3. Defineix una funció que, donada una frase, retorni el nombre de paraules úniques que conté sense fer servir la funció len(). (Pots utilitzar el mètode split() d’un string.)

```python
# Definim una funció que compta quantes paraules úniques hi ha en una frase
def comptar_paraules_uniques(frase):
    # Convertim la frase a minúscules i la separem en paraules
    # Això evita comptar com a diferents "Casa" i "casa"
    paraules = frase.lower().split()  

    # Inicialitzem una llista buida per guardar paraules que ja hem trobat
    paraules_uniques = []

    # Comptador per saber quantes paraules úniques hi ha
    comptador = 0

    # Recorrem totes les paraules de la frase
    for i in paraules:
        # Si la paraula no està a la llista de paraules úniques, l’afegim
        if i not in paraules_uniques:
            paraules_uniques.append(i)
            comptador += 1  # Incrementem el comptador de paraules úniques

    # Retornem el nombre total de paraules úniques
    return comptador

# Demanem a l’usuari que introdueixi una frase
entrada = input("Introdueix una frase: ")

# Cridem la funció i guardem el resultat
nombre_uniques = comptar_paraules_uniques(entrada)

# Mostrem el resultat a l’usuari
print(f"La frase conté {nombre_uniques} paraules úniques.")
```

5) Escriu una funció que demani un text i retorni un diccionari amb les lletres que apareixen i la quantitat de cops que apareixen, sense fer servir funcions d’agregació com count().

```python
# Definim la funció per comptar les lletres
def comptar_lletres():
    # Demanem al/la usuari/ària que introdueixi un text
    text = input("Introdueix un text: ")
   
    # Inicialitzem un diccionari buit per guardar les lletres i la seva freqüència
    lletres = {}

    # Recorrem cada caràcter del text
    for caracter in text:
        # OPCIONAL: Comprovem si el caràcter és una lletra (podem filtrar símbols o espais)
        if caracter.isalpha():
            # Convertim la lletra a minúscula per comptar sense distinció de majúscules
            lletra = caracter.lower()

            # Si la lletra ja existeix al diccionari, incrementem el seu valor
            if lletra in lletres:
                lletres[lletra] += 1
            else:
                # Si és la primera vegada que apareix, l'afegim amb valor 1
                lletres[lletra] = 1

    # Retornem el diccionari amb les freqüències
    return lletres

# Cridem la funció i guardem el resultat
resultat = comptar_lletres()

# Imprimim el diccionari resultant
print("Freqüència de les lletres:")
print(resultat)
```

6) Crea una funció que demani 5 noms de plats i el seu nombre de calories. A mesura que s’introdueixen, guarda aquestes dades en un diccionari i el retorni. Executa la funció i imprimeix el diccionari resultant.

```python
# Definim una funció que demana 5 plats amb les seves calories
def demanar_plats():
    plats = {}  # Diccionari buit per guardar els plats i les calories

    # Bucle per demanar 5 plats
    for i in range(5):
        nom = input(f"Introdueix el nom del plat {i+1}: ")
        calories = int(input(f"Introdueix el nombre de calories del plat '{nom}': "))
        plats[nom] = calories  # Guardem la dada al diccionari

    # Retornem el diccionari complet
    return plats

# Cridem la funció i guardem el resultat
diccionari_plats = demanar_plats()

# Imprimim el diccionari resultant
print("\nDiccionari de plats i calories:")
print(diccionari_plats)
```

#### <span style="color: rgb(230, 0, 126);"> Exercici 1 (2p)</span>

Crea una funció que rebi una llista de paraules com a paràmetre. 
La funció ha de recórrer la llista i retornar una llista amb la paraula més llarga a la primera posició i la paraula més curta a la segona posició. 
Després, mostra per pantalla les dues paraules.

Fes servir aquesta llista per executar la funció:
arbre, codi, helicòpter, opcions, sal, amanida, cotxe, sofà, llapis, pissarra

*Fes l'exercici sense utilitzar funcions de python com max() o min().*

```python
#escriu el teu codi python
def contador(llistat):
    llista_resultant=[]
    maxim=llistat[0] #por el momento nuestro maximo es la primera palabra, haré que el bucle la vaya actualizando y lo mismo con el minimo
    minim=llistat[0]

    for paraula in llistat:
        if len(paraula)>len(maxim):
            maxim=paraula #actualizamos el maximo cada vez que encuentra uno superior
        if len(paraula)<len(minim):
            minim=paraula 

    llista_resultant.append(maxim)
    llista_resultant.append(minim)

    return llista_resultant

llistat=['arbre', 'codi', 'helicòpter', 'opcions', 'sal', 'amanida', 'cotxe', 'sofà', 'llapis', 'pissarra']
seleccion=contador(llistat)
print(f'La paraula més llarga i la més curta han estat: {seleccion}')
```

#### <span style="color: rgb(230, 0, 126);"> Exercici 2 (2p)</span>

Crea una funció que demani 5 noms d'alumnes i les seves respectives notes de finals de curs.

A mesura que vagis demanant les dades, has d’afegir:
- Al diccionari d’aprovats, les persones que tenen una nota igual o superior a 5.
- Al diccionari de suspesos, la resta d’alumnes.

La funció ha de retornar el diccionari d'aprovats.

Després, mostra per pantalla el diccionari d’aprovats.

Fes servir aquestes dades per executar la funció:
Laura - 8 || Antoni - 10  || Laia - 4  ||  Albert - 2 || Sara - 9

```python
#escriu el teu codi python
def evaluador():
    aprovats = {}
    suspesos = {}
    for alumne in range(5):
        nom = input(f'Introdueix el nom del alumne número {alumne+1}: ') #interpreto que es la funcion la que debe ir pidiendo ronda a ronda el alumno y su nota
        nota = int(input(f'Introdueix la nota del alumne {nom}: '))
        if nota >= 5:
            aprovats[nom] = nota
        else:
            suspesos[nom] = nota
    return aprovats

evaluacio=evaluador()
print(f'Aquests són els alumnes que han aprovat: {evaluacio}')
```

#### <span style="color: rgb(230, 0, 126);"> Exercici 4 (3p)</span>

En una empresa, és essencial avaluar el rendiment dels venedors per optimitzar les vendes i millorar els resultats. Un indicador clau és comparar les vendes de cada venedor amb la mitjana de l’equip comercial.

Escriu una funció que donat el dataset vendes.csv:

- Agrupi les vendes per venedor i calculi el total de cada un.
- Calculi la mitjana de vendes de tots els venedors.
- Identifiqui els venedors amb un rendiment inferior a la mitjana.
- Retorni una llista amb els noms dels venedors que han tingut vendes inferiors a la mitjana.

Crida la funció amb el dataframe vendes.csv i mostra els venedors que han tingut un rendiment en vendes per sota de la mitjana.


```python
#escriu el teu codi python
def rendiment(df):
    vendes=df.groupby('Venedor')[['Import']].sum()
    mitjana=vendes['Import'].mean() #Utilizo vendes['Import'] porque interpreto que debemos hacer la media de vendes de los 5 vendedores, no la de vendedor individual ni la media de importe de todos los productos.
    no_eficients=[]
    for venedor, fila in vendes.iterrows(): #con iterrows() iteramos fila a fila y no en columnas
        if fila['Import']<mitjana:
            no_eficients.append(venedor)
        else:
            pass
    return no_eficients

import pandas as pd
df=pd.read_csv('/Users/noeliasolano/python/vendes.csv', sep=';')

rendiment(df)
```
